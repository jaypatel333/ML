```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters = 20,max_iter = 100 ) #20 fire turcks, limiting maximum iteration
# default = (8,300) 
kmeans = kmeans.fit(mat) # Fitting the input data
# Getting the cluster labels
#labels = kmeans.predict(mat)

centroids = kmeans.cluster_centers_ # Centroid values
```
